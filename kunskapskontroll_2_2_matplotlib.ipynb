### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Plotting can be made by using an (1) object oriented approach - by explicitely creating figures, axes and use methods on them, or by (2) implicitely using pyplot to create and manage the content and the plotting.

It's recommended to use the (1) object oriented approach - especially for more complicated plots, whereas (2) pyplot can be good to use when you quickly needs a plot.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

(1) The Figure object is the canvas where a plot is created - The size of figure can be adjusted, and the plots are made by creating axes which are child of figure.

(2) The Axes object are a child of figure, and is the region in a figure that a plot is plotted. With axes methods you can configure what to plot, adjust axis scales, add labels and more. One figure can have several axes - which enables you to plot several plots.

(3) The axis object is a component of Axes, which can set limits for the axes, determine scale of ticks, create ticklabels and more.

(4) Artist is basically all the building blocks that you can see on a figure - The figure itself, axes, axis, text and other objects are part of this.

### When plotting in Matplotlib, what is the expected input data type?

The expected input is numpy arrays (or masked arrays - subclass of ndarray).
Similar array structures can be used but might not work as you want.
The overall convention is to convert data into ndarray before plotting.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object oriented approach
x = np.linspace(-4, 4, 1000)
fig, ax = plt.subplots(figsize=(7, 4), layout='constrained')
ax.plot(x, x**2, label='x^2')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Figure created with OO-approach')
ax.legend()

# Implicit approach with pyplot
plt.figure(figsize=(7, 4), layout='constrained')
plt.plot(x, x**2, label='x^2', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Figure made with pyplot')
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
#random_data_x = np.random.randn(1000)
#random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Creates figure for our two plots
fig, axs = plt.subplots(1,2, figsize=(10,8))

#First - We plot the x^2 function
axs[0].plot(x, y, label = "f(x)")
axs[0].set_title("f(x) = x^2")
axs[0].set_xlabel("X-values")
axs[0].set_ylabel("Y-values")
axs[0].legend()

#Secondly - We plot the bar plot
axs[1].bar(names, values)
axs[1].set_title("Bar plot over fruits")
axs[1].set_xlabel("Type of fruit")
axs[1].set_ylabel("Total amount")
```
